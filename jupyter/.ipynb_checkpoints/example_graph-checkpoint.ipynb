```python
import pandas as pd
# matplotlib : 메트플롯 반드시 import 한다
import matplotlib.pyplot as plt
# 숫자 계산을 쉽게 하기 위한 라이브러리
import numpy as np
# 데이터 사이를 부드럽게 연결해주는 함수
from scipy.interpolate import make_interp_spline

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

#### 과목별 점수를 세로 막대그래프로 표현하시오
#### 그래프 제목: "과목별 점수" 
#### x축: 과목 / y축: 점수 
#### 막대 위에 점수 값 표시
#### 아래 데이터를 이용해 작성하시오.

#### + python 항목 부분에만 레이블 값이 출력되도록 수정

```python
df = pd.DataFrame({
    "과목": ["Java", "Python", "React", "SQL"],
    "점수": [85, 90, 75, 80]
})

plt.figure(figsize=(5, 4))

bar01 = plt.bar(df['과목'], df['점수'], width=0.5, color='#8c5ee8')

for i in range(len(bar01)) :
    bar = bar01[i]
    height = bar.get_height()

    if df['과목'][i] == 'Python' :
            plt.text(
                bar.get_x() + bar.get_width() / 2, 
                height,
                int(height), 
                ha='center',
                va='bottom'
            )

plt.title('과목별 점수')
plt.xlabel('과목')
plt.ylabel('점수')

plt.show()
```

#### 공부시간을 가로 막대그래프로 표현하시오
#### 막대 두께를 0.5로 설정
#### 그래프 제목: "학생별 공부시간"
#### 아래 데이터를 이용해 작성하시오.
#### + 김학생 제외하기

```python
df = pd.DataFrame({
    "이름": ["김학생", "이학생", "박학생", "최학생"],
    "공부시간": [3, 5, 2, 4]
})

plt.figure(figsize=(5, 3))

df = df[df['이름'] != '김학생']
bar02 = plt.barh(df['이름'], df['공부시간'], color='#8c5ee8', height=0.5)

plt.title('학생별 공부시간')
plt.xlabel('공부시간')
plt.ylabel('이름')

plt.show()
```

#### 매출 데이터를 선그래프로 표현
#### marker 추가
#### 선 아래 영역을 색으로 채우기
#### 그래프 제목: "월별 매출 변화" 
#### 아래 데이터를 이용해 작성하시오.

```python
df = pd.DataFrame({
    "월": ["1월", "2월", "3월", "4월"],
    "매출": [100, 150, 130, 170]
})

bar03 = plt.plot(df['월'], df['매출'], color='#8c5ee8', marker='*')

plt.fill_between(df['월'], df['매출'], alpha=0.1, color='#8c5ee8')

plt.title('월별 매출 변화')
plt.xlabel('월')
plt.ylabel('매출')

plt.show()
```

#### 사용자 비율을 도넛 그래프로 표현 
#### 가운데 원 추가 
#### 그래프 제목: "OS 점유율" 
#### 아래 데이터를 이용해 작성하시오.

```python
df = pd.DataFrame({
    "OS": ["Windows", "Mac", "Linux"],
    "사용자수": [70, 20, 10]
})

plt.figure(figsize=(4,4))

plt.pie(df['사용자수'], labels=df['OS'], autopct='%1.1f%%', startangle=90)

cir = plt.Circle((0,0), 0.3, fc='white')
plt.gca().add_artist(cir)

plt.title('OS 점유율')

plt.show()
```

#### 아래 두개의 데이터를 병합하여 작성하시오.
#### 결측치는 각 컬럼의 평균값으로 처리하시오
#### 기본급과 성과급, 수당을 이용하여 누적 세로 막대그래프를 생성하세요
#### 그래프 구성
#### 아래 : 기본급 / 성과급(누적)
#### 범례 추가 
#### 그래프 제목 : 부서별 급여 구조
#### x축 : 부서 / y축 : 금액
#### 그래프를 이미지로 저장 : salary_chart.png, dpi=300
#### 아래 세개의 데이터를 병합하여 작성하시오 

```python
df1 = pd.DataFrame({
    "부서": ["개발팀", "영업팀", "인사팀"],
    "기본급": [300, 250, 200]
})

df2 = pd.DataFrame({
    "부서": ["개발팀", "영업팀", "마케팅팀"],
    "성과급": [100, 80, 70]
})

df3 = pd.DataFrame({
    "부서": ["개발팀", "영업팀", "마케팅팀"],
    "수당": [50, 40, 30]
})

plt.figure(figsize=(4,4))

df = pd.merge(df1, df2, how='outer', on='부서')
df = pd.merge(df3, df, how='outer', on='부서')

df['기본급'] = df['기본급'].fillna(df['기본급'].mean())
df['성과급'] = df['성과급'].fillna(df['성과급'].mean())
df['수당'] = df['수당'].fillna(df['수당'].mean())

# plt.bar(df['부서'], df['기본급'], bottom=df['성과급'], label='기본급', width=0.5)
# plt.bar(df['부서'], df['성과급'],  bottom=df['수당'], label='성과급', width=0.5)
# plt.bar(df['부서'], df['수당'], label='수당', width=0.5)
plt.bar(df['부서'], df['기본급'], label='기본급', width=0.5)
plt.bar(df['부서'], df['성과급'], bottom=df['기본급'], label='성과급', width=0.5)
plt.bar(df['부서'], df['수당'], bottom=df['기본급']+df['성과급'], label='수당', width=0.5)

plt.title('부서별 급여 구조')
plt.xlabel('부서')
plt.ylabel('금액')

plt.legend()

plt.show()
```
