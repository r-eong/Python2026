```python
dataFile = open('greenData/test_data.txt', 'r', encoding='utf-8 sig')
dataLine = dataFile.readlines()
dataLine  # 출력
# dataFile.close()  # 출력안됨
```

Readline() : 오픈한 파일 디스크립터.readline() 를 호출하여 현재까지 읽은 파일 데이터의 다음 한 줄을 문자열타입으로 읽음

```python
dataFile = open('greenData/test_data.txt', 'r', encoding='utf-8 sig')
dataLine = dataFile.readline()
print(dataLine)
```

Read() : 오픈 파일 디스크립터.read() 를 호출해서 전체 파일 데이터를 문자열 타입으로 읽음 .
# with 는 읽고, close 까지 한꺼번에 블럭으로 묶어서 처리해줌.
# f 는 방금 연 그 파일 통로를 앞으로 이 코드 안에서는 f라고 부르겠다고 이름을 붙인 것.
# open() 은 파일을 여는 동작을 수행한 뒤, 그 파일을 다룰 수 있는 통로를 결과물로 돌려준다.

```python
with open('greenData/test_data.txt', 'r', encoding='utf-8 sig') as f:
    content = f.read()
    print(content)
```

문제) score_data.csv 파일의 데이터중 국어점수가 85점 이상인 데이터만 추출하여 출력하세요.

```python
import csv
# DictReader : csv 파일을 읽기 좋게 변환하는데 첫 줄을 제목(=필드명)으로 인식하도록 변경해줌.

# 1. 파일 읽기
with open('greenData/score_data.csv', 'r', encoding='utf-8 sig') as f:
    # 2. csv 의 첫 줄을 필드명으로 변경
    reader = csv.DictReader(f)

    print('국어점수가 85점 이상인 데이터 : ')
    
    # 3. 한 줄씩 확인해서 if 로 조건 검사
    for r in reader:
        # 파일에서 읽어온 모든 숫자는 문자열이므로 반드시 int 로 변환한다.
        if int(r['국어']) >= 85:
            print(r)
```
