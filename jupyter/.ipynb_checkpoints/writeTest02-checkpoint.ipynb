```python
from bs4 import BeautifulSoup

# 파일 읽기
with open('greenData/member_data.xml', 'r', encoding= 'utf-8 sig') as f:
    # 파싱
    soup = BeautifulSoup(f, 'xml')

# 데이터 추출 (select 모두 출력. 내가 지정한 'member'(반드시 태그여야함)안의 모든 것)
members = soup.select('member')

for m in members :
    name = m.select_one('name').text
    age = m.select_one('age').text
    city = m.select_one('city').text

    print(f"이름 : {name} | 나이 : {age} | 도시 : {city}")
```

```python
# pandas 라이브러리 : 테이블형 데이터를 다룰 수 있는 다양한 기능을 가진 라이브러리
# pandas 는 반드시 import 해야함
# row data : 아직 데이터 분석을 위해 정제되지 않은 기본데이터
# 데이터를 다루는 방식 2가지
#   1. 시리즈(serise)는 1차원(행 하나 혹은 열 하나) 데이터
#   2. 데이터 프레임(Dataframe)은 2차원(테이블형) 데이터

import pandas as pd  # 라이브러리 임포트 (대부분 pd 로 사용)
# pd.판다스 함수 형태로 호출하여 사용

# index 는 행의 레이블(=제목)을 의미
# index 를 지정하지 않으면 자동으로 0부터 생성
# 지정할 경우 지정된 index 부터 생성

# seriesData = pd.Series([70, 80, 90])
seriesData = pd.Series([70, 80, 90], index=['국어', '영어','수학'])
seriesData

seriesData.index  # index 만 출력

# Series 수정
seriesData.index = ['미술','사회','과학']
seriesData

# Series 삭제
del seriesData['사회']
seriesData
```

```python
# pandas 데이터 타입
# object : String 또는 혼용 데이터(문자열)
# int64 : 파이썬의 int(정수)와 같음
# float64 : 파이썬의 float(부동 소수점)과 같음
# bool : 파이썬의 bool 또는 boolean(True/False)과 같음
# datetitme64 : 날짜/시간
# timedelta : 두개의 datetime64 간의 차이 예) 2026-02-22 와 2026-03-01 둘의 간격 계산

seriesData = pd.Series(['dava', 'alex', 'amir'])
seriesData

seriesData = pd.Series([1, 2, 3])
seriesData

# int64 의 정수를 float64 부동소수점으로 변경가능
# Series.astype(변경할 타입)
seriesData.astype('float')

# 데이터 타입 확인
seriesData = pd.Series(['dava', 1, 2, 3])
seriesData

seriesData = pd.Series([True, False, True])
seriesData
```

```python
df = pd.DataFrame({
    "미국": [2.1, 2.2, 2.3],
    "한국": [0.4, 0.5, 0.45],
    "중국": [10, 13, 15]
}, 
    index = [2020, 2021, 2022]
)

# 데이터프레임 데이터 접근하기
# df.loc : index 를 통해서 값을 찾는다.
# df.iloc : index 번호를 통해서 값을 찾는다. 0부터 시작

df.loc[2020]
# index 라는 데이터프레임의 레이블을 이용해 값을 출력
# 미국     2.1
# 한국     0.4
# 중국    10.0
# Name: 2020, dtype: float64

df.iloc[1]
# 이미 가지고 있는 0부터 시작하는 index 로 값을 출력
# 미국     2.2
# 한국     0.5
# 중국    13.0
# Name: 2021, dtype: float64

# 특정 값 추출할 때
# df['한국'][2020]
# np.float64(0.4)
print(df['한국'][2020])
# 0.4

# 실제로 지워지는 건 아님! 보이는 것만 삭제되는 것!
# 데이터프레임에 컬럼(=열=새로) 삭제 : del
del df['미국']
df
#    	한국	중국
# 2020	0.40	10
# 2021	0.50	13
# 2022	0.45	15

# 데이터프레임에 행(=가로) 삭제 :drop([인덱스 이름])
df.drop([2020])
# 	   미국	한국 중국
# 2021  2.2	0.50  13
# 2022	2.3	0.45  15
```

```python
import pandas as pd

data = {
    '대여소': ['강남', '강남', '홍대', '홍대', '잠실', '잠실'],
    '이용자수': [120, 150, 80, 90, 200, 180],
    '대여시간': [30, 45, 20, 25, 60, 55],
    '이용일자': ['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02']
}

df = pd.DataFrame(data)
```

```python
# Data 자료를 이용해 이용자수가 100명 이상인 데이터만 출력하시오.
df[df['이용자수'] >= 100]
```

```python
# Data 자료를 이용해 대여시간 기준으로 오름차순 정렬하여 출력하시오.
# sort_values 기본값은 오름차순 df.sort_values(by='대여시간')
# 내림차순 df.sort_values(by='대여시간', ascending = False)
df.sort_values(by='대여시간')
```

```python
# Data 자료를 이용해 "총이용량" 컬럼을 이용자수 * 대여시간으로 계산하여 추가하시오.
df['총이용량'] = df['이용자수'] * df['대여시간']
df
```

```python
# Data 자료를 이용해 컬대여소별 총 이용자수 구하여 출력하시오.
df.groupby('대여소')['이용자수'].sum()
```

```python
# Data 자료를 이용해 대여소별 총 이용자수를 구한 후 이용자수 많은 순으로 정렬하여 출력하시오.
df.groupby('대여소')['이용자수'].sum().sort_values(ascending = False)
```

```python
# Data 자료를 이용해 이용자수가 100 이상인 데이터만 추출하여 대여소별 평균대여시간 구하여 출력하시오.
df[df['이용자수'] >= 100].groupby('대여소')['대여시간'].mean()
```

```python
# Data 자료를 이용해 대여소별 "총이용량" 합계구하기
df.groupby('대여소')['총이용량'].sum()
```

```python
# 이용 일자별 이용자수 합계가 가장 많은 날짜 출력
busyDate = df.groupby('이용일자')['이용자수'].sum().sort_values(ascending = False)
print(busyDate)
print(f"가장 바쁜 날짜 : {busyDate.index[0]}")
```

```python
# 이용자수 평균이 가장 높은 대여소 찾기.
manyFind = df.groupby('대여소')['이용자수'].mean().sort_values(ascending = False)
print(manyFind)
print(f"평균이 가장 높은 대여소 : {manyFind.max()}")
```
