```python
# 병합(merge) : 최소 2개 이상의 데이터프레임을 합치는 작업
import pandas as pd

pd.set_option('display.max_columns', None)  # 모든 컬럼 출력
pd.set_option('display.width', 1000)  # 출력 너비 지정
pd.set_option('display.unicode.east_asian_width', True)  # 모든 컬럼 출력

# 예시 데이터
df1 = pd.DataFrame({
    "지역": ["서울", "부산", "대구"],
    "총인구": [1000000, 300000, 200000]
})
df2 = pd.DataFrame({
    "지역": ["서울", "부산", "인천"],
    "병원수": [150, 80, 60]
})

print(f"\n필드명 확인")

# 데이텊레임의 필드명 확인 함수 : 데이터프레임이름.columns
print(df1.columns)
print(df2.columns)
# Index(['지역', '총인구'], dtype='object')
# Index(['지역', '병원수'], dtype='object')

print(f"\nmerge")

# pd.merge(left_데이터프레임_이름, right_데이터프레임_이름, how="타입", on="기준컬럼")
# how 사용방식 : inner(교집합), left(왼쪽기준), right(오른쪽 기준), outer(합집합)

merged01 = pd.merge(df1, df2, how='inner', on='지역')
print(merged01)
#    지역   총인구  병원수
# 0  서울  1000000     150
# 1  부산   300000      80

print(f"\n합집합")

# 합집합
merged02 = pd.merge(df1, df2, how='outer', on='지역')
print(merged02)
#    지역     총인구  병원수
# 0  대구   200000.0     NaN
# 1  부산   300000.0    80.0
# 2  서울  1000000.0   150.0
# 3  인천        NaN    60.0

print(f"\nleft")

# left
merged03 = pd.merge(df1, df2, how='left', on='지역')
print(merged03)
#    지역   총인구  병원수
# 0  서울  1000000   150.0
# 1  부산   300000    80.0
# 2  대구   200000     NaN

print(f"\nright")

# right
merged06 = pd.merge(df1, df2, how='right', on='지역')
print(merged06)
#    지역     총인구  병원수
# 0  서울  1000000.0     150
# 1  부산   300000.0      80
# 2  인천        NaN      60

print()
print("-------------------------------------")
print(f"\n컬럼명(필드명)이 다른 경우")

# 컬럼명(필드명)이 다른 경우
df3 = pd.DataFrame({
    "지역": ["서울", "부산", "대구"],
    "총인구": [1000000, 300000, 200000]
})
df4 = pd.DataFrame({
    "지역명": ["서울", "부산", "인천"],
    "병원수": [150, 80, 60]
})

merged04 = pd.merge(df3, df4, how='inner', left_on='지역', right_on='지역명')
print(merged04)
#    지역   총인구 지역명  병원수
# 0  서울  1000000   서울     150
# 1  부산   300000   부산      80

print()
print("-------------------------------------")
print(f"\n여러개의 컬럼명으로 병합하는 경우")

# 여러개의 컬럼명으로 병합하는 경우
df5 = pd.DataFrame({
    "지역": ["서울", "부산", "대구"],
    "연도": [2024, 2025, 2024],
    "총인구": [1000000, 300000, 200000]
})
df6 = pd.DataFrame({
    "지역": ["서울", "부산", "인천"],
    "연도": [2024, 2025, 2024],
    "병원수": [150, 80, 60]
})
merged05 = pd.merge(df5, df6, how='inner', on=['지역', '연도'])
print(merged05)
#    지역  연도   총인구  병원수
# 0  서울  2024  1000000     150
# 1  부산  2025   300000      80
```

```python
# 결측치 : 데이터 병합하는 과정에서 데이터가 비어있는 상태를 의미
# 즉, 값이 없어서 NaN(Not a Number) 로 나타나는 경우

df1 = pd.DataFrame({
    "지역": ["서울", "부산", "대구"],
    "총인구": [1000000, 300000, 200000]
})
df2 = pd.DataFrame({
    "지역": ["서울", "부산", "인천"],
    "병원수": [150, 80, 60]
})

print(f"\n결측치 변환후 병합")
# 병합
mer01 = pd.merge(df1, df2, how='outer', left_on='지역', right_on='지역')
print(mer01)
#    지역     총인구  병원수
# 0  대구   200000.0     NaN
# 1  부산   300000.0    80.0
# 2  서울  1000000.0   150.0
# 3  인천        NaN    60.0

print(f"\n결측 개수 확인")
# 결측 개수 확인
# isna : 결측치 존재
print(mer01.isna().sum())
# 지역      0
# 총인구    1
# 병원수    1
# dtype: int64

print(f"\n------- 결측치 처리 -------")
# 결측치 처리방법
# 1. 특정 값으로 채우기 : filla(0 혹은 평균값...)
# 총 인구 결측치를 0으로 채우기
print(f"\nfilla - 0으로 채우기")
# mer01['총인구'] = mer01['총인구'].fillna(0)
# mer01['병원수'] = mer01['병원수'].fillna(0)
# print(mer01)
#    지역     총인구  병원수
# 0  대구   200000.0     0.0
# 1  부산   300000.0    80.0
# 2  서울  1000000.0   150.0
# 3  인천        0.0    60.0

print(f"\nfilla - 평균값으로 채우기")

mer01['총인구'] = mer01['총인구'].fillna(mer01['총인구'].mean())
mer01['병원수'] = mer01['병원수'].fillna(mer01['병원수'].mean())
mer01['비율'] = mer01['총인구'] / mer01['병원수']
mer01.style.format({'총인구': '{:,.0f}', '병원수':'{:.0f}', '비율':'{:.1f}'})
#  	지역	총인구   	병원수	비율
# 0	대구	200,000   	97	    2069.0
# 1	부산	300,000	    80   	3750.0
# 2	서울	1,000,000	150	    6666.7
# 3	인천	500,000	    60  	8333.3

# 2. 행 삭제하기 : dropna
# dropna 는 결측이 존재하는 행을 통째로 삭제
# print(f"\ndropna")
# mer01 = mer01.dropna()
# print(mer01)
#    지역     총인구  병원수
# 0  대구   200000.0     0.0
# 1  부산   300000.0    80.0
# 2  서울  1000000.0   150.0
# 3  인천        0.0    60.0


```
