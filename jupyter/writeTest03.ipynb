```python
# 미세먼지 데이터 분석
# 일시	구분	미세먼지(PM10)	초미세먼지(PM25)
```

```python
import pandas as pd

df = pd.read_csv('greenData/pm10.csv', encoding= 'cp949')
# 구분컬럼의데이터중'평균'인행을제외하여실제자치구데이터만남깁니다. (데이터전처리)
print('--- 미세먼지 및 초미세먼지 주의 지역 ---')
# df.drop(df['미세먼지(PM10)'].mean())  # 틀림 
disc = df[df['구분'] != '평균']

# 미세먼지(PM10) 농도가 25 이상이면서 동시에초미세먼지(PM25) 농도가 15 이상인 '주의 요망' 
# df[df['미세먼지(PM10)']>=25]
# df[df['초미세먼지(PM25)']>= 15]
condition = (disc['미세먼지(PM10)'] >= 25) & (disc['초미세먼지(PM25)'] >= 15)
condi02 = disc[condition]

# 조건에맞는지역의구분, 미세먼지(PM10), 초미세먼지(PM25) 컬럼만 선택해 출력하세요.
# del df['일시']  # 좋지않은 방법이라함!
# df
result = condi02[['구분', '미세먼지(PM10)', '초미세먼지(PM25)']]
print(result)
```

```python
import pandas as pd

df = pd.read_csv('greenData/accident.csv', encoding= 'cp949')

# pandas 에서 날짜를 추출하는 함수 종류
# csv 에 입력된 모든 날짜 및 숫자는 문자열 데이터 타입이다.
# 날짜형식으로 형변환하는 함수 : datetime()
# 사고일자의 월 출력
# 년 : df.year, 월 : df.month, 일 : df.day, 시 : df.hour, 분 : df.minute, 초 : second
# 요일을 숫자로 반환 : df.dayofweek (0:원 ~ 6일)
# 요일 이름이 문자열 ('Monday') : df.day_name()
# 분기(1분기~4분기) : df.quarter

# 사고일자 컬럼을 datetime 형식으로 변환하고, 이를 활용하여 '사고월'이라는 새로운 파생변수(컬럼)를 
# 생성하세요.
print('--- 월별 사고 발생 건수 ---')
df['사고일자'] = pd.to_datetime(df['사고일자'])
df['사고월'] = df['사고일자'].dt.month

# 월별 사고건수를 집계하여 어느월에 사고가 가장 많이 발생했는지 확인하세요.
# groupby().count() : 데이터가 비어있지 않은 값이 개수를 센다
# broupby().size() : 그룹별 전체 행의 개수를 센다. (비어있어도 셈)

sago = df.groupby('사고월')['사고월'].count()
# ┖> 빈값이 들어있으면 빈 값의 행을 무시하기때문에 빈 값을 세고싶다면 size
# sago = df.groupby('사고월').size()
sagoMonth = sago.idxmax()  # idxmax : 가장 큰 값이 들어있는 위치(인덱스 이름)에 접근하여 가져옴

print(sago)
print(f"사고가 가장 많은 월 : {sagoMonth}월", end=' / ')
print(f"발생 건수 : {sago.max()}건")

print()

# 사망자수와 부상자수를 더하여 '사상자수'라는 파생변수를 생성하세요.
print('--- 사상자수 기준 상위 5개 위험 노선 ---')
df['사상자수'] = df['사망'] + df['부상']

# 노선명별로 그룹화하여 '사상자수'의 총합을 구하고, 사상자가 가장 많이 발생한 상위5개 노선을
# 내림차순으로 출력하세요.
result = df.groupby('노선명')['사상자수'].sum().sort_values(ascending = False).head(5)
print(result)

# (보너스) 사고 원인이'졸음'이면서사망자가1명이상발생한치명적인사고는총몇건인지구하세요.
print('--- 졸은운전 중 사망자가 발생한 치명적 사고 건수 ---')
result02 = len(df[(df['원인'] == '졸음') & (df['사망'] >= 1)])
print(f"{result02}건")
```
